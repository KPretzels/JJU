## 유사도 결과
(안녕하세요? 반갑습니다! <=====> Hi, nice to meet you.)

| Model                                             | 유사도 결과 |
|---------------------------------------------------|-------------------|
| sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2 | 0.7905 |
| sentence-transformers/xlm-r-100langs-bert-base-nli-stsb-mean-tokens | 0.8110 |
| sentence-transformers/paraphrase-xlm-r-multilingual-v1 | 0.6974 |
| sentence-transformers/LaBSE | 0.7024 |
| sentence-transformers/distiluse-base-multilingual-cased-v2 | 0.7696 |
| sentence-transformers/paraphrase-multilingual-mpnet-base-v2 | 0.8560 |
| sentence-transformers/stsb-xlm-r-multilingual | 0.8110 |




```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# SentenceTransformer로 다국어 지원 임베딩 모델 불러오기
model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')

# 문장 정의
sentences = [
    "안녕하세요? 반갑습니다.",
    "안녕하세요? 반갑습니다!",
    "안녕하세요? 만나서 반가워요.",
    "Hi, nice to meet you.",
    "I like to eat apples."
]

# 문장 임베딩
embedded_sentences = model.encode(sentences)

# 유사도 계산 함수
def similarity(a, b):
    return cosine_similarity([a], [b])[0][0]

# 문장 간 유사도 계산 및 출력
for i, sentence in enumerate(embedded_sentences):
    for j, other_sentence in enumerate(embedded_sentences):
        if i < j:
            print(f"[유사도 {similarity(sentence, other_sentence):.4f}] {sentences[i]} \t <=====> {sentences[j]}")
```
