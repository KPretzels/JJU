```python
with open('/content/appendix-keywords.txt') as f:
  file = f.read()
```

```python
print(file[:5])
```

```python
# CharacterTextSplitter
```

```python
from langchain.text_splitter import CharacterTextSplitter, RecursiveCharacterTextSplitter
```

```python
char_splitter = CharacterTextSplitter(chunk_size=250, chunk_overlap=50)
char_chunks = char_splitter.split_text(file)[:10]
```

```python
char_chunks[1]
```

```python
# RecursiveCharacterTextSplitter
```

```python
recursive_splitter = RecursiveCharacterTextSplitter(chunk_size=250, chunk_overlap=50)
recursive_chunks = recursive_splitter.split_text(file)[:10]
```

```python
recursive_chunks[1]
```
