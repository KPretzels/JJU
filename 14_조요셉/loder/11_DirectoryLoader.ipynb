```python
from langchain_community.document_loaders import DirectoryLoader

loader = DirectoryLoader('./', glob='**/*.md') # 셈플데이터로 출력함.
docs = loader.load()
len(docs)
```

```python
print(docs[0].page_content[:1000])
```

# MutiThread 사용법

```python
from langchain_community.document_loaders import DirectoryLoader

loader = DirectoryLoader('./', glob='**/*.md', use_multithreading=True) # 셈플데이터로 출력함.
docs = loader.load()
len(docs)
```

# Loader Class

```python
from langchain_community.document_loaders import TextLoader

loader = DirectoryLoader('./', glob='**/*.md', loader_cls=TextLoader) # 셈플데이터로 출력함.
docs = loader.load()
len(docs)
```

```python
print(docs[0].page_content)
```
